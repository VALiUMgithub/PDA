```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.ar_model import AR
df = pd.read_csv("Downloads\stock_data.csv",
parse_dates=True,
index_col="Date")
# displaying the first five rows of dataset
df.head()
```

```python
df.drop(columns='Unnamed: 0')
df['Volume'].plot()
```

```python
df.plot(subplots=True, figsize=(4,4))
```

```python
df_month = df.resample("M").mean()
# using subplot
fig, ax = plt.subplots(figsize=(6, 6))
# plotting bar graph
ax.bar(df_month['2016':].index,
df_month.loc['2016':, "Volume"],
width=25, align='center')
```

```python
df.Low.diff(2).plot(figsize=(6,6))
```

```python
df.High.diff(2).plot(figsize=(10,6))
```

```python
window_size=50
rolling_mean = df['Open'].rolling\
(window_size).mean()
rolling_mean.plot()
```
