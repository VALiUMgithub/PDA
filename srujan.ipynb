```python
import pandas as pd
data = {'Name': ['Jai', 'Prince', 'Gaurav', 'Anuj', 'Ravi', 'Natasha', 'Riya'],
        'Age': [17, 18, 20, 19, 21, 18, 19],
        'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M'],
        'Marks': [93, 95, 82, 75, 'NaN', 64, 'NaN']}
df = pd.DataFrame(data)
df
```

```python
c = avg = 0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c += 1
        avg += ele
avg /= c
df = df.replace(to_replace = "NaN", value = avg)
print(avg)
df
```

```python
df['Gender'] = df['Gender'].map({'M': 0, 'F': 1}).astype(float)
df
```

```python
df = df[df['Marks'] >= 70].copy()
df
```
